```
from rdkit import Chem
from rdkit.Chem import Draw
```

# The top 3 drugs

```
M = Chem.MolFromSmiles('O=C(O)CC(=O)NCC(c1ccccc1)C(O)CC(Cc1ccccc1)NC(=O)OC1CCCCC1')
m = Chem.MolFromSmiles('CNC(=O)C(Cc1ccccc1)NC(=O)C(Cc1ccccc1)NC(=O)C(Cc1ccccc1)NC(=O)C(CC(N)=O)NC(=O)c1ccccc1') 
Mole = Chem.MolFromSmiles('Cc1ccccc1C(=O)NCCc1ccccc1C(=O)NCCc1ccccc1C(=O)NCCc1ccccc1C(=O)') 
```

```
# Visualizing the SMILES
# The numbers represent the Docking Scores/Binding Affinities of the top 3 drugs
Draw.MolsToGridImage([M, m, Mole], 
                     legends=["-6.7", "-6.4", "-6.0"])
```
